```python
from numpy import random, array
```

```python
def createclusterdata(P,k):
    random.seed(10)
    points_per_cluster=float(P)/k
    x = []
    for i in range(k):
        income_centroid=random.uniform(20000.0,200000.0)
        age_centroid=random.uniform(20.0,80.0)
        
        for j in range(int(points_per_cluster)):
            x.append([random.normal(income_centroid,10000.0),random.normal(age_centroid,5.0)])
            
    x=array(x)
    return x
        
    
```

```python
%matplotlib inline

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from numpy import random, float
```

```python
cluster= createclusterdata (1000,8)
model = KMeans(n_clusters=8)
model = model.fit(scale(cluster))
print(model.labels_)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(cluster[:,0], cluster[:,1], c=model.labels_.astype(float))
```
